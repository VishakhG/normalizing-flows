```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import pyro
import pyro.optim
import pyro.distributions as dist
%config Completer.use_jedi = False
%matplotlib inline
```

```python
def w_1(z0, z1):
    return np.sin((2 * np.pi * z0) / 4)

def w_2(z0, z1):
    return 3 * np.exp(-.5 * ((z0 - 1) / .6) ** 2)

def sigma(x):
    return 1 / (1 + np.exp(- x))

def w_3(z0, z1):
    return 3 * sigma((z0 - 1) / .3) 

def norm(x, y):
    return np.linalg.norm(np.array([x, y]))

def pot_1(z0, z1):
    outer_term_1 = .5 * ( (norm(z0, z1) - 2) / .4 ) ** 2
    inner_term_1 = np.exp( (-.5 * ((z1 - 2) / .6 ) ** 2) )
    inner_term_2 = np.exp( (-.5 * ((z1 + 2) / .6 ) ** 2) )
    outer_term_2 = np.log(inner_term_1 + inner_term_2)
    return outer_term_1 - outer_term_2

def pot_2(z0, z1):
    return .5 * ((z1 - w_1(z0, z1)) / .4 ) ** 2


def pot_3(z0, z1):
    term_1 = np.exp(-.5 * ((z1 - w_1(z0, z1)) / .35) **2) 
    term_2 = np.exp(-.5 * ((z1 - w_1(z0, z1) + w_2(z0, z1)) / .35) ** 2)
    return -np.log(term_1 + term_2)

def pot_4(z0, z1):
    term_1 = np.exp(-.5 * ((z1 - w_1(z0, z1)) / .4) **2) 
    term_2 = np.exp(-.5 * ((z1 - w_1(z0, z1) + w_3(z0, z1)) / .35) ** 2)
    return -np.log(term_1 + term_2)

def plot_pot_func(pot_func, ax, sparse=False, transpose = False):
    x = np.linspace(-4, 4, 1e2)
    y = np.linspace(-4, 4, 1e2)
    xx, yy = np.meshgrid(x, y, sparse=sparse)
    z = np.exp(- pot_func(xx, yy))
    cmap =  plt.get_cmap('inferno')
    
    if transpose:
        z = z.T
    ax.contourf(x, y, z, cmap=cmap)
    
def generate_data(pot_func):
    x = np.linspace(-4, 4, 1e2)
    y = np.linspace(-4, 4, 1e2)
    xx, yy = np.meshgrid(x, y, sparse=sparse)
    z = np.exp(- pot_func(xx, yy))
    
    return z 
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 20))
axes = axes.flat

plot_pot_func(pot_1, axes[0], True, True)
plot_pot_func(pot_2, axes[1], False, False)
plot_pot_func(pot_3, axes[2], False, False)
plot_pot_func(pot_4, axes[3], False, False)
```

# torchifying

```python
def w_1(z):
    return np.sin((2 * np.pi * z[0]) / 4)

def w_2(z):
    return 3 * np.exp(-.5 * ((z[0] - 1) / .6) ** 2)

def sigma(x):
    return 1 / (1 + np.exp(- x))

def w_3(z):
    return 3 * sigma((z[0] - 1) / .3) 

def pot_1(z):
    outer_term_1 = .5 * ( (z.norm() - 2) / .4 ) ** 2
    inner_term_1 = torch.exp( (-.5 * ((z[1] - 2) / .6 ) ** 2) )
    inner_term_2 = torch.exp( (-.5 * ((z[1] + 2) / .6 ) ** 2) )
    outer_term_2 = torch.log(inner_term_1 + inner_term_2)
    return outer_term_1 - outer_term_2

def pot_2(z):
    return .5 * ((z[1] - w_1(z) / .4 )) ** 2


def pot_3(z):
    term_1 = torch.exp(-.5 * ((z[1] - w_1(z[0], z[1])) / .35) **2) 
    term_2 = np.exp(-.5 * ((z[1] - w_1(z) + w_2(z)) / .35) ** 2)
    return -np.log(term_1 + term_2)

def pot_4(z):
    term_1 = torch.exp(-.5 * ((z[1] - w_1(z)) / .4) **2) 
    term_2 = torch.exp(-.5 * ((z[1] - w_1(z) + w_3(z)) / .35) ** 2)
    return -torch.log(term_1 + term_2)


class pot_1_pdf(pyro.distributions.TorchDistribution):
    def __init__(self):
        super(pot_1_pdf, self)
    def log_prob(self, value):
        return torch.exp(- pot_1(value))



x = np.arange(-4, 4, 1)
y = np.arange(-4, 4, 1)
g = np.meshgrid(x, y)
positions = np.vstack(map(np.ravel, g))
positions.shape
```

```python
def model(data):
    for i in range(len(data)):
        pyro.sample("obs_{}".format(i), pot_1_pdf(), obs=data[i])

def model():
    pyro.sample("obs".format(i), pot_1_pdf())
```

# Simple guide


```python
def simple_guide(data):
    a = pyro.param("a", torch.tensor([0.0, 0.0]))
    b = pyro.param("b", torch.tensor(1.))
    res = []
    for i in range(len(data)):
        res.append(pyro.sample("obs_{}".format(i), dist.Normal(a, b)))
    return res
```

```python
def simple_guide():
    a = pyro.param("a", torch.tensor([0.0, 0.0]))
    b = pyro.param("b", torch.tensor(1.))
    return pyro.sample("obs", dist.Normal(a, b))
```

```python
svi = pyro.infer.SVI(model=model,
                     guide=simple_guide,
                     num_samples=10,
                     optim=pyro.optim.Adam({"lr": 1e-5}),
                     loss=pyro.infer.Trace_ELBO().differentiable_loss)
losses = []
num_steps = 10
for i in range(num_steps):
    losses.append(svi.step())
plt
```

```python

pyro.clear_param_store()

svi = pyro.infer.SVI(model=model,
                     guide=simple_guide,
                     num_samples=10,
                     optim=pyro.optim.Adam({"lr": 1e-5}),
                     loss=pyro.infer.Trace_ELBO().differentiable_loss)


losses, a,b  = [], [], []
num_steps = 2500
data = torch.Tensor(positions.T)
for t in range(num_steps):
    losses.append(svi.step(data))
    a.append(pyro.param("a"))
    b.append(pyro.param("b"))

plt.plot(losses)
plt.title("ELBO")
plt.xlabel("step")
plt.ylabel("loss");
print('a =a ',pyro.param("a"))
print('b = ', pyro.param("b"))
```

```python
simple_guide(data)
```

```python
posterior=svi.run(data)
trace_pred = pyro.infer.TracePredictive(model,
                             posterior,
                             num_samples=1000)
post_pred = trace_pred.run(data)

marg = pyro.infer.EmpiricalMarginal(post_pred, ['obs_{}'.format(i) for i in range(len(data))])

import seaborn as sns
sampled = marg.sample().numpy()
cmap =  plt.get_cmap('inferno')

sns.jointplot(sampled[:, 0], sampled[:, 1], kind='kde', cmap=cmap)
```

# Planar flow

```python
def planar_guide(data):
    a = pyro.param("a", torch.tensor([0.0, 0.0]))
    b = pyro.param("b", torch.tensor(1.))
    base_dist = dist.Normal(a, b)
    plfs = [dist.PlanarFlow(2)] * 2
    plf_modules = [pyro.module("my_plf_{}".format(i), plfs[i]) for i in range(len(plfs))]

    plf_dist = dist.TransformedDistribution(base_dist, plfs)
    res = []
    for i in range(len(data)):
        res.append(pyro.sample("obs{}".format(i), plf_dist))
    return res




pyro.clear_param_store()
svi = pyro.infer.SVI(model=model,
                     guide=planar_guide,
                      num_samples=10,
                     optim=pyro.optim.RMSprop({"lr": 1e-5, "momentum":.1}),
                     loss=pyro.infer.Trace_ELBO())


losses, a,b  = [], [], []
num_steps = 10000
data = torch.Tensor(positions.T)
for t in range(num_steps):
    losses.append(svi.step(data))
    a.append(pyro.param("a"))
    b.append(pyro.param("b"))

plt.plot(losses)
plt.title("ELBO")
plt.xlabel("step")
plt.ylabel("loss");
print('a =a ',pyro.param("b"))
print('b = ', pyro.param("a"))

```

```python

sampled = np.vstack([np.array([elem.data.numpy() for elem in planar_guide(data)]) for _ in range(1)])
cmap =  plt.get_cmap('inferno')

sns.jointplot(sampled[:, 0], sampled[:, 1], kind='kde', cmap=cmap)

```
